```python
import os, glob, cv2, math, csv, tqdm
import numpy as np
import util 
from keras.utils.training_utils import multi_gpu_model
from keras import backend as k
from keras import applications, optimizers, losses, layers
from keras.utils.training_utils import multi_gpu_model
from keras import layers ,losses
from keras.layers import Dropout, Input, Flatten, Dense, GlobalAveragePooling2D, Activation, LSTM, Reshape, TimeDistributed, GRU
from keras.models import Sequential, Model, load_model
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
from keras.layers.convolutional import Conv2D
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def test_generator(root, seq, n_frame, batch):
    seq_pose = seq +'.txt'
    img_dir = os.path.join(root, 'resize_img', seq)
    lab_dir = os.path.join(root, 'poses', seq_pose)
    img_list = sorted(os.listdir(img_dir))
    lab_list = util.load_dof6_labels(lab_dir)
    idx = n_frame
    while 1:
        img = []
        lab = []
        
        for i in range(batch+1):
            bat_img = []
            bat_lab = []
            #print(idx)
            img_seq_list = img_list[idx:idx+n_frame]
            bat_lab = lab_list[idx:idx+n_frame]
            for img_name in img_seq_list:
                img_path = os.path.join(img_dir, img_name)
                tmp_img = cv2.resize(cv2.imread(img_path),(512,384))/255.
                bat_img.append(tmp_img)
            img.append(np.array(bat_img))
            lab.append(np.array(bat_lab))
        print(idx)
        idx = idx + 1
        yield np.concatenate((np.array(img)[1:], np.array(img)[:-1]),axis = 4), [np.array(lab)[1:,:,:3],np.array(lab)[1:,:,3:]]
```

```python
def data_generator(root, seq, n_frame, batch):
    while 1:
        seq_ch = np.random.choice(seq)
        seq_pose = seq_ch +'.txt'
        img_dir = os.path.join(root, 'resize_img', seq_ch)
        lab_dir = os.path.join(root, 'poses', seq_pose)
        img_list = sorted(os.listdir(img_dir))
        lab_list = util.load_dof6_labels(lab_dir)
        img = []
        lab = []        
        for i in range(batch+1):
            bat_img = []
            bat_lab = []
            idx = np.random.randint(0, len(img_list)-n_frame)
            #print(idx)
            img_seq_list = img_list[idx:idx+n_frame]
            bat_lab = lab_list[idx:idx+n_frame]
            for img_name in img_seq_list:
                img_path = os.path.join(img_dir, img_name)
                tmp_img =  cv2.resize(cv2.imread(img_path),(512,384))/255.
                bat_img.append(tmp_img)
            img.append(np.array(bat_img))
            lab.append(np.array(bat_lab))
        #print(seq_ch)    
        yield np.concatenate((np.array(img)[1:], np.array(img)[:-1]),axis = 4), [np.array(lab)[1:,:,:3],np.array(lab)[1:,:,3:]]
```

```python
n_time = 20
batch = 4

train_seq = ['01','03','04','06']
train_root= ''
train_generator = data_generator(train_root, train_seq, n_time, batch)
```

```python
model_gen = load_model('../GAN/save_gen.h5')
model_gen.summary()
model_top = Model(inputs = model_gen.input, 
                  outputs=model_gen.layers[9].output)
```

```python
inputs = Input(shape=(None, 384, 512,6))
x = TimeDistributed(model_top)(inputs)
x = TimeDistributed(GlobalAveragePooling2D())(x)
x = GRU(1024, return_sequences=True)(x)
x = GRU(1024, return_sequences=True)(x)
rotation = Dense(512)(x)
rotation = Dense(3, name = 'rotation')(rotation)
position = Dense(512)(x)
position = Dense(3, name = 'position')(position)
```

```python
model = Model(inputs = inputs, outputs = [rotation, position])
for i in range(3):
    model.layers[i].trainable = False
model.compile(optimizer='adam', loss=['mse','mse'],loss_weights=[100,1])
```

```python
model.summary()
```

```python
hist = model.fit_generator(train_generator, steps_per_epoch=1000, epochs=1)
```

```python
#plt.plot(hist.history['loss'],'r-')
plt.plot(hist.history['rotation_loss'],'b-')
plt.plot(hist.history['position_loss'],'g-')
plt.show()
```
